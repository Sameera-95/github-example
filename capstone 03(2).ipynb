# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

# Scrape from webpage

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
# print(soup.prettify())
```

```python
rows = soup.find_all('tr')
```

```python
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
```

```python
df = pd.DataFrame(list_rows)
df
```

# Cleaning the data frame

```python
df1 = df[0].str.split(',', expand=True)
df1.head()
```

```python
df1[0] = df1[0].str.strip('[')
df1[0] = df1[0].str.strip('\n')
df1[1] = df1[1].str.strip('\n')
df1[2] = df1[2].str.strip('\n]')
df1.head(10)
```

```python
df2 = df1.iloc[:,0:3]
df2
```

```python
col_labels = soup.find_all('th')
all_header = []
col_str = str(col_labels)
cleantext2 = BeautifulSoup(col_str, "lxml").get_text()
all_header.append(cleantext2)
print(all_header)
```

```python
df2 = pd.DataFrame(all_header)
df2.head()
```

```python
df3 = df2[0].str.split(',', expand=True)
df3.head()
```

```python
frames = [df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
df4[0] = df4[0].str.strip('[')
```

```python
df5 = df4.rename(columns=df4.iloc[0])
df5.head()
```

```python
df6 = df5.iloc[:,0:3]
df6
```

```python
df6.dtypes
```

```python
df6.columns =["Postal_code" , "Borough", "Neighborhood"] 
```

```python
df6.dtypes
```

# Removing missing and unwanted values

```python
df7 = df6.drop(df6[df6.Neighborhood == " "].index)
df7
```

```python
df8 = df7.drop(df7[~ ( df7['Postal_code'].str.startswith("M") ) ].index)
df8
```
